# Student Course Management System

## Project Notebook

### Project Overview
This project is a Django-based Student Course Management System that allows students to enroll in courses and admins to manage courses.

### Tech Stack
- Python
- Django
- SQLite
- HTML, CSS
- GitHub

### Features
- User Registration & Login
- Course Listing
- Student Enrollment
- Admin Dashboard

### Project Structure
- courses app
- templates
- SQLite database
- Django admin

### How to Run
1. Clone repository
2. Create virtual environment
3. Install requirements
4. Run `python manage.py runserver`

### Conclusion
This project helped me understand Django MVT architecture and backend development.
